```python
import pandas as pd
import glob
```

```python
glob.glob('./res_gossiping/*.txt')
```

```python
data = list()
for filePath in glob.glob('./res_gossiping/*.txt'):
    with open(filePath, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('---split---')[1].split('\n')
        data.append(tmpList[1:-1])
data
```

```python
columns = [c.split(': ')[0] for c in data[0]]
columns
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def columnFilter(s):
    output = s.split(': ')[1]
    return output
```

```python
df['推'].apply(columnFilter)
```

```python
df2 = df.copy()
```

```python
for c in df:
    print(c)
```

```python
for c in df:
    df[c] = df[c].apply(columnFilter)
```
